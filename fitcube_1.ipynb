```python
%matplotlib inline
import matplotlib.pyplot as plt
from eddy.fit_annulus import annulus
from eddy.modelling import gaussian_ensemble
import numpy as np
from eddy import rotationmap
np.random.seed(123)
```

```python
ensemble = gaussian_ensemble(vrot=1500., Tb=40., dV=350., rms=2.0, N=20, plot=True, return_annulus=True)
```

```python
# Get the percentiles of the parameters.
vrot, *hyperparams = ensemble.get_vlos(fit_method='GP')
print("Best fit rotation velocity was %.1f +\- %.1f m/s." % (vrot[1], 0.5 * (vrot[2] - vrot[0])))
```

```python
annulus = gaussian_ensemble(vrot=1500., Tb=40., dV=350., rms=2.0, N=20, tau=10.0, plot=True, return_annulus=True)
vrot, *hyperparams = annulus.get_vlos(fit_method='GP')
print("Best fit rotation velocity was %.1f +\- %.1f m/s." % (vrot[1], 0.5 * (vrot[2] - vrot[0])))
```
